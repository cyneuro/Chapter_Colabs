## A1. Concept cards — start with the membrane
**A1.1 Semi‑permeable membrane & selective ion channels**  \
Neuronal membranes contain **ion‑selective channels** embedded in a lipid bilayer; permeability to a given ion depends on whether its channels are open and their conductance. Selectivity (e.g., Na⁺ vs K⁺ vs Cl⁻) is the basis for how ionic gradients create membrane currents. *Read:* Dayan & Abbott, Ch. 5–6; Koch, Ch. 1 & 4. citeturn24search34turn24search137

**A1.2 Nernst potential (equilibrium potential)**  \
For a **single permeant ion**, the membrane voltage at which **chemical** and **electrical** driving forces balance is given by the **Nernst equation**; at this voltage, net current for that ion is zero. *Read:* Dayan & Abbott; Koch (derivations and intuition). citeturn24search34turn24search137

**A1.3 Ion‑specific reversal potentials**  \
Each ion (Na⁺, K⁺, Cl⁻, etc.) has its **own reversal (equilibrium) potential** determined by its **inside/outside concentrations** via the Nernst relation. In current‑clamp traces, synaptic events drive Vm toward the **E_rev** of the underlying conductance. *Read:* Dayan & Abbott; Koch. citeturn24search34turn24search137

**A1.4 Nernst = reversal (for a selective channel)**  \
For a channel that is effectively **selective to a single ion**, the **Nernst potential for that ion is the channel’s reversal potential** (the voltage at which the current through that channel changes sign). *Read:* Dayan & Abbott; Koch. citeturn24search34turn24search137

**A1.5 Membrane RC & τ (recap)**  \
Low‑pass behavior; step → exponential; brief inputs approximate an impulse response. *Read:* Dayan & Abbott; Koch. citeturn24search34turn24search137

**A1.6 Cable & λ (recap)**  \
Distance‑dependent attenuation/delay along dendrites; implications for distal vs proximal input influence. *Read:* Rall; Stuart & Spruston. citeturn24search93turn24search74

**A1.7 Spike mechanism (qualitative)**  \
Voltage‑gated Na⁺/K⁺ dynamics generate threshold and refractoriness. *Read:* Hodgkin–Huxley (1952). citeturn24search148

**A1.8 E/I timing motifs**  \
Feedforward inhibition clamps peak depolarization and sharpens temporal fidelity. *Read:* Pouille & Scanziani (Science, 2001). citeturn24search42

**A1.9 Core concepts (program outcomes)**  \
Structure–function, information flow, plasticity; cross‑discipline adoption of core ideas. *Read:* SfN/BrainFacts; AAAS Vision & Change; NASEM STEM report. citeturn26search1
